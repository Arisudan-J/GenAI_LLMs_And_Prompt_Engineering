```python

from google.colab import userdata
from groq import Groq
import json
api_key = userdata.get("API_KEY")
client = Groq(api_key=api_key)
product_name = input("Enter a product name: ")
prompt = f"""
Generate information about the product '{product_name}'.

Return ONLY valid JSON with the following fields:
- name
- description
- price
- category

Do not include any explanation or markdown formatting.
"""
response = client.chat.completions.create(
      model="llama-3.3-70b-versatile",
          response_format={"type": "json_object"},
              messages=[
                      {
                                  "role": "user",
                                              "content": prompt
                                                                }
                                                          ]
                                                          )
response_text = response.choices[0].message.content



```

```python
print("Raw API Response:")
response_text
```

```python
product_data = json.loads(response_text)

print("\nPython Dictionary:")
product_data
```

```python
required_fields = [
      "name",
          "description",
              "price",
                  "category"
                  ]

try:
  product_data = json.loads(response_text)

  for field in required_fields:
    if field not in product_data:
      raise ValueError(f"Missing field: {field}")

  print("JSON is valid.")
  print(product_data)
except json.JSONDecodeError:
  print("Error: Invalid JSON received.")

except ValueError as e:
  print("Validation Error:", e)

except Exception as e:
  print("Unexpected Error:", e)
```

```python
improved_prompt = f"""
Generate details for the product '{product_name}'.

Rules:
1. Return ONLY valid JSON.
2. Do not include markdown.
3. Do not include extra text.
4. Include exactly these fields:
{{
  "name": "",
  "description": "",
  "price": 0,
  "category": ""
}}
5. Price must be a number.
"""
```

```python
response = client.chat.completions.create(
      model="llama-3.3-70b-versatile",
          response_format={"type": "json_object"},
              messages=[
                      {
                                  "role": "user",
                                              "content": improved_prompt
                                                      }
                                                          ]
                                                          )

improved_response = response.choices[0].message.content
improved_response


```

```python
product_data = json.loads(improved_response)
print("\nImproved JSON:")
product_data
```
