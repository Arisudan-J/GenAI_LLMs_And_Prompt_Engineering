```python
from google.colab import userdata
api_key=userdata.get('GEMINI_API_KEY')
```

```python
from google import genai

client = genai.Client(api_key=api_key)

email_type = input("Enter email type (leave request/complaint): ").strip().lower()
```

```python
if email_type == "leave request":
    details = input("Enter leave reason and days: ")

    prompt = f""" Write a professional leave request email. Details: {details} """


elif email_type == "complaint":
    details = input("Enter complaint details: ")

    prompt = f""" Write a professional complaint email.Details: {details}"""


elif email_type == "meeting request":
    details = input("Enter meeting purpose and time: ")

    prompt = f""" Write a professional meeting request email.Details: {details}"""


else:
    prompt = "Write a professional email."

```

```python
response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents=prompt
)
```

```python
email_type = input("Enter email type (leave request/complaint/meeting request): ").strip().lower()
```

```python
if email_type == "leave request":
    details = input("Enter leave reason and days: ")

    prompt = f"""
Write a professional leave request email.

Details: {details}

Requirements:
- Add a subject line.
- Use a formal tone.
- Keep it under 120 words.
- End with a professional closing.
"""


elif email_type == "complaint":
    details = input("Enter complaint details: ")

    prompt = f"""
Write a professional complaint email.

Details: {details}

Requirements:
- Add a subject line.
- Use a polite and formal tone.
- Keep it under 120 words.
- Request a solution.
"""


elif email_type == "meeting request":
    details = input("Enter meeting purpose and time: ")

    prompt = f"""
Write a professional meeting request email.

Details: {details}

Requirements:
- Add a subject line.
- Use a formal tone.
- Keep it under 120 words.
- End with a professional closing.
"""


else:
    prompt = """
Write a professional email.

Requirements:
- Add a subject line.
- Use a formal tone.
- Keep it under 120 words.
"""
```

```python
response1 = client.models.generate_content(
    model="gemini-3.5-flash",
    contents=prompt
)
```
